```python
import numpy as np
import pandas as pd # we use pandas library to analyse and structure the data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to the pandas dataframe
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head() #first 5 rows of dataset
```

```python
#in dataset,output  0-good transaction,1-fraud transaction
credit_card_data.tail()
```

```python
credit_card_data.info() # dataset information
```

```python
credit_card_data.isnull().sum() #checking the missing values in each column
```

```python
credit_card_data['Class'].value_counts() # checking ow many valid and how many are fraud
```

```python
# this dataset is highly unbalanced dataset
```

```python
legal = credit_card_data[credit_card_data.Class == 0] # all the rows having output 0 is stored in this variable
fraud = credit_card_data[credit_card_data.Class == 1] # all the rows having output 1 is stored in this variable
print(legal.shape)
print(fraud.shape)
```

```python
legal.Amount.describe() # statistical info of data
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean() # compare the values for both transactions
```

```python
# Here it is unbalanced data set,so we have to do Under Sampling
# we have to Build a sample dataset containing similar distribution of normal transactions and fraud transactions
# fraud transactions = 492
# now i take some random 492 legal transactions and i will merge with 492 fraud transactions,now it is a balanced dataset.
```

```python
legal_sample =legal.sample(n= 492)
new_dataset =pd.concat([legal_sample,fraud],axis=0) # concatenating 2 datsets to make a overall balanced dataset
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into Features and Target
X = new_dataset.drop(columns='Class',axis=1) # dropping the last column
Y = new_dataset['Class']
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.2,stratify =Y,random_state=2) # stratify is to divide Y equally in train and test data
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model Training

# Logistic Regression
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)

#Training the Logistic Regression Model with Training data
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Model Evaluation

# accuracy on training data
X_train_prediction =model.predict(X_train) # by using model we are predicting the training data
training_data_accuracy = accuracy_score(X_train_prediction,Y_train) # calculating the accuracy
print('Accuracy on Training Data : ',training_data_accuracy)
```

```python
# it is a good score
# if the prediction is more than 70%,then we can say it is well trained
# This model is having Low Bias
```

```python
# accuracy for test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on Test Data : ',testing_data_accuracy)
```

```python
# it is a good score
# This model is having Low Variance

# This model is a Best Fit for Logistic Regression
```

```python
X_new = X_test.iloc[[0]]

prediction = model.predict(X_new)
print(prediction)

if(prediction[0] == 0):
  print('Credit Card Transaction is Legal')
else:
  print('Credit Card Transaction is Fraud')
```
